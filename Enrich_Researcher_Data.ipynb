This notebook is used to enrich the csv file containing the researcher data, in order to suit our needs for lab 3.

```python
import pandas as pd
from faker import Faker
```

```python
df = pd.read_csv("/Users/alfredo.leon/Desktop/SDMLab1/PropertyGraphsLab/neo4jcontainerdata/dataset.csv", encoding='latin1')
df.head(n=3).T
```

```python
from random import choice, sample, randint
import random
random.seed(0)
Faker.seed(0)
fak = Faker()
Faker.seed(0)

# Assign random paper type to the rows
paper_types = ["Short Paper", "Full Paper", "Demo Paper", "Poster"]
df['PaperType'] = 0
df['PaperType'] = df.PaperType.apply(lambda x: choice(paper_types))

# Assign random conference type to the rows
conferences = df['General conference name'].unique()
conference_type = ['Workshop', 'Symposium', 'ExpertGroup', 'RegularConference']
conferences_types = {conference: choice(conference_type) for conference in conferences}
df['Conference Type'] = df['General conference name'].apply(lambda x: conferences_types[x])

# Assign chairs or editors
leaders = [fak.name() for _ in range(20)]
conferences_leaders = {conference: choice(leaders) for conference in conferences}
df['Leader'] = df['General conference name'].apply(lambda x: conferences_leaders[x])

# Assign reviewers to the papers
reviewers = [fak.name() for _ in range(20)]
n_reviewers = df.Title.apply(lambda x: randint(2, 4))
df["Reviewers"] = n_reviewers.apply(lambda x: sample(reviewers, x))

# Assign reviews to the papers
df["Review_decisions"] = n_reviewers.apply(lambda x: [choice(['Accepted', 'Rejected']) for _ in range(x)])

# Assign review comments
df["Review_comments"] = n_reviewers.apply(lambda x: [fak.paragraph(nb_sentences=5) for _ in range(x)])

df.head().T

```
